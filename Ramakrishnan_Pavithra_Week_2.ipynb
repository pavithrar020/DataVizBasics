```python
import numpy as np

# Creating a 4-dimensional array
arr4d = np.array([[[[1, 3, 5, 7], [2, 4, 6, 8]]]])

# Creating a 2-dimensional array
arr2d = np.array([[10, 20, 30], [40, 50, 60]])

# Printing the shape of arrays
print("4D Array Shape:", arr4d.shape)
print("2D Array Shape:", arr2d.shape)
```

```python
from matplotlib import pyplot as plt

data = np.zeros((512, 512, 3), dtype=np.uint8)
data[256, 256] = [255, 0, 0]
```

```python
# Display with a different color map 
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.title("Different Color Map")
plt.show()
```

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt

# display an image with a red patch
w_pil, h_pil = 512, 512
data_pil = np.zeros((h_pil, w_pil, 3), dtype=np.uint8)
data_pil[0:256, 0:256] = [255, 0, 0]  # red patch in upper left
img_pil = Image.fromarray(data_pil, 'RGB')
img_pil.show()
```

```python
# display an image
w_np, h_np = 512, 512 #width and height
data_np = np.zeros((h_np, w_np, 3), dtype=np.uint8)
data_np[[256, 256]] = [255, 0, 0] # coordinates
data_np[[200, 200]] = [255, 0, 0] # coordinates

# new method used to display image
plt.imshow(data_np, interpolation='nearest')
plt.title("Image with Nearest Interpolation")
plt.show()
```

```python
X = arr2d[0, :]  # Using the first row of arr2d as X values
Y = arr2d[1, :]  # Using the second row of arr2d as Y values

# Plotting points using scatter method
plt.scatter(X, Y)
plt.title("Scatter Plot from 2D Array")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# coordinates of the parallelogram
x = [1, 4, 5, 2, 1]
y = [1, 2, 5, 4, 1]

# Plot parallelogram with different color lines
plt.plot(x[:2], y[:2], color='blue', linestyle='-', marker='o', markersize=8)
plt.plot([x[1], x[2]], [y[1], y[2]], color='green', linestyle='-', marker='o', markersize=8)
plt.plot(x[2:], y[2:], color='red', linestyle='-', marker='o', markersize=8)
plt.plot([x[3], x[0]], [y[3], y[0]], color='purple', linestyle='-', marker='o', markersize=8)

# Set aspect ratio to be equal
plt.gca().set_aspect('equal', adjustable='box')

plt.title("Parallelogram with Different Color Lines")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([10, 30, 25, 35])
labels = ['Blue', 'Green', 'Teal', 'Purple']

# Peacock colors: Blue, Green, Teal, Purple
colors = ['#005e7e', '#00a591', '#6ab187', '#a87caa']

plt.pie(y, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show() 
```
